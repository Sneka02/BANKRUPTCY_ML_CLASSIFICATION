```python
#DEFINING DEPENDENT AND INDEPENDENT VARIABLES
x = dataset.drop('Bankrupt?', axis=1)
y = dataset['Bankrupt?']
```

```python
#SPLITTING THE DATA INTO TRAINING AND TESTING SET
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
#APPLYING PCA
from sklearn.decomposition import PCA
n_components = 2
pca = PCA(n_components=n_components)
pca.fit(x_train)
pc_train = pd.DataFrame(pca.transform(x_train), columns=["PC" + str(i+1) for i in range(n_components)])
pc_test = pd.DataFrame(pca.transform(x_test), columns=["PC" + str(i+1) for i in range(n_components)])
pc_train
```

```python
from imblearn.over_sampling import SMOTE
```

```python
sm = SMOTE(random_state = 2)
x_train_res, y_train_res = sm.fit_resample(x_train, y_train)
```

```python
sm = SMOTE(random_state=2)
x_test_res, y_test_res = sm.fit_resample(x_test, y_test)
```

```python
y_train_res.value_counts()
```

```python
y_test_res.value_counts()
```

```python
#FEATURE SCALING
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train_res = sc.fit_transform(x_train_res)
x_test_res = sc.fit_transform(x_test_res)
```

```python
#RANDOM FOREST CLASSIFIER
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators = 60, criterion = 'entropy',random_state = 0)
rf.fit(x_train_res,y_train_res)
```

```python
#GETTING CONFUSION MATRIX
from sklearn.metrics import confusion_matrix,classification_report
y_pred = rf.predict(x_test_res)
cr = classification_report(y_test_res,y_pred)
cm = confusion_matrix(y_test_res,y_pred)
print('confusion matrix:\n',cm)
```

```python
pca_model = RandomForestClassifier()
pca_model.fit(pc_train, y_train)
print("PCA : {:3f}".format(pca_model.score(pc_test, y_test)*100))
```

```python
import seaborn as Sns
sns.heatmap(confusion_matrix(y_test_res,y_pred),annot=True,cmap='Blues',fmt='g')
```

```python
#CHECKING THE ACCURACY OF THE MODEL
from sklearn.metrics import accuracy_score
dta = accuracy_score(y_test_res,y_pred)
print('accuracy score = ',accuracy_score(y_test_res,y_pred))
print("Classification Report",classification_report(y_test_res,y_pred))
```

```python
#DECISIONTREE CLASSIFIER
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(criterion = 'entropy')
dt.fit(x_train_res,y_train_res)
```

```python
#CONFUSION MATRIX
from sklearn.metrics import confusion_matrix,classification_report
y_pred = dt.predict(x_test_res)
cm = confusion_matrix(y_test_res,y_pred)
print('confusion matrix:\n',cm)
```

```python
pca_model = DecisionTreeClassifier()
pca_model.fit(pc_train, y_train)
print("PCA : {:3f}".format(pca_model.score(pc_test, y_test)*100))
```
